```python
# Define ROC function
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

def draw_roc(actual, probs):
    fpr, tpr, thresholds = roc_curve(actual, probs)
    roc_auc = auc(fpr, tpr)
    
    plt.plot(fpr, tpr)
    plt.plot([0,1],[0,1],'r--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.show()
    
    return roc_auc
```

```python
# Call the ROC function
draw_roc(y_train_final, y_train_pred_final['Attrition_Left_Prob'])
```

```python
# Predict on training data by creating columns with different probability cutoffs to explore the impact of cutoff on model performance
numbers = [float(x)/10 for x in range(10)]

for i in numbers:
    y_train_pred_final[i] = y_train_pred_final['Attrition_Left_Prob'].map(lambda x: 1 if x > i else 0)

y_train_pred_final.head()
```

```python
# Create a DataFrame to see the values of accuracy, sensitivity, and specificity at different values of probability cutoffs
cutoff_df = pd.DataFrame(columns=['prob','accuracy','sensi','speci'])

for i in numbers:
    cm = metrics.confusion_matrix(y_train_pred_final['Attrition_Left'], y_train_pred_final[i])
    TP = cm[1,1]
    TN = cm[0,0]
    FP = cm[0,1]
    FN = cm[1,0]
    
    cutoff_df.loc[len(cutoff_df)] = [
        i,
        (TP+TN)/(TP+TN+FP+FN),
        TP/(TP+FN),
        TN/(TN+FP)
    ]

cutoff_df
```

```python
# Plot accuracy, sensitivity, and specificity at different values of probability cutoffs
cutoff_df.plot(x='prob', y=['accuracy','sensi','speci'])
plt.show()
```
